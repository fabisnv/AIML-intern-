```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
df['Cabin'] = df['Cabin'].fillna('Unknown')
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
df = pd.get_dummies(df, columns=['Sex', 'Embarked'], drop_first=True)

display(df.head())
```

```python
df.info()
```

```python
df.describe()
```

```python
from sklearn.preprocessing import StandardScaler

numerical_cols = ['Age', 'Fare']


scaler = StandardScaler()
df[numerical_cols] = scaler.fit_transform(df[numerical_cols])
```

```python
df.head()
```

```python
df.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


numerical_cols = ['Age', 'SibSp', 'Parch', 'Fare']

plt.figure(figsize=(10, 6))
for i, col in enumerate(numerical_cols):
    plt.subplot(2, 2, i + 1)
    sns.boxplot(y=df[col])
    plt.title(col)

plt.tight_layout()
plt.show()
```

```python

numerical_cols = ['Age', 'SibSp', 'Parch', 'Fare']


Q1 = df[numerical_cols].quantile(0.25)
Q3 = df[numerical_cols].quantile(0.75)
IQR = Q3 - Q1

df_cleaned = df[~((df[numerical_cols] < (Q1 - 1.5 * IQR)) |(df[numerical_cols] > (Q3 + 1.5 * IQR))).any(axis=1)]


print("Original DataFrame shape:", df.shape)
print("Cleaned DataFrame shape:", df_cleaned.shape)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


numerical_cols = ['Age', 'SibSp', 'Parch', 'Fare']

plt.figure(figsize=(10, 6))
for i, col in enumerate(numerical_cols):
    plt.subplot(2, 2, i + 1)
    sns.boxplot(y=df[col])
    plt.title(col)

plt.tight_layout()
plt.show()
```

Preprocessing Steps Summary:
Data Loading and Exploration: Loaded the dataset and examined its structure, missing values, and data types.
1.Handling Missing Values: Imputed missing values in 'Age' with the mean, 'Embarked' with the mode, and 'Cabin' with a placeholder.
2.Categorical Feature Conversion: Converted 'Sex' and 'Embarked' into numerical representations using one-hot encoding (or label encoding as requested later).
3.Numerical Feature Scaling: Standardized 'Age' and 'Fare' using StandardScaler.
4.Outlier Visualization and Removal: Visualized outliers using boxplots and removed them using the IQR method.
5.The dataset df_cleaned now contains the preprocessed data with missing values handled, categorical features encoded, numerical features scaled, and outliers removed.
